```python
import numpy as np
import random
def coordenadas_humano(orientaciones,all_init_cords,barcos):
    barco = random.choice(barcos)
    init_cord = random.choice(all_init_cords) 
    orientacion  = random.choice(orientaciones) 

    x_init, y_init = init_cord

    if orientacion == 'N':
        coords_barco = list(zip(list(range(x_init - barco, x_init, 1)), [y_init] * barco))
    elif orientacion == 'S':
        coords_barco = list(zip(list(range(x_init, x_init + barco, 1)), [y_init] * barco))
    elif orientacion == 'E':
        coords_barco = list(zip([x_init]* barco, list(range(y_init, y_init + barco, 1))))  
    else:
        coords_barco =list(zip([x_init]* barco, list(range( y_init - barco, y_init, 1))))

    return coords_barco, barco, init_cord, orientacion


def within_range_humano(tablero, orientacion, barco, init_coord):
    x, y = tablero.shape
    x_init, y_init = init_coord

    if orientacion == 'N':
        return (x_init - barco) >= 0
    elif orientacion == 'S':
        return (x_init + barco) <= x
    elif orientacion == 'E':
        return (y_init + barco) <= y
    else:
        return (y_init - barco) >= 0
def adyacentes(coordenada):
    x, y = coordenada
    adyacentes = [(x-1, y-1), (x-1, y), (x-1, y+1), (x, y-1), (x, y+1), (x+1, y-1), (x+1, y), (x+1, y+1)]
    return adyacentes

def intersec_humano(tablero,coords_barco):
    contexto = np.where(tablero == 'O')
    contexto_tablero = list(zip(contexto[0], contexto[1]))
    for coord in coords_barco:
        adyacentes_coord = adyacentes(coord)
        for adyacente in adyacentes_coord:
            if adyacente in contexto_tablero:
                return True
    return False
tablero_juego = np.full((10,10),' ')
x, y = tablero_juego.shape
x_axis = np.arange(x)
y_axis = np.arange(y)
allppartida = [(p1, p2) for p1 in x_axis for p2 in y_axis]
puntos_cardinales_def = ['N', 'S', 'E', 'O']
barqui単os = [1, 1, 1, 1, 2, 2, 2, 3, 3, 4]
barcos_coordenadas_humano = []


def colocar_barco_humano(tablero, buques, todas_coordenadas, puntos_cardinales):

    coordenadas_barco, fragata, punto_partida, pto_cardinal  = coordenadas_humano(puntos_cardinales, todas_coordenadas, buques)
    dentro_rango = within_range_humano(tablero, pto_cardinal, fragata, punto_partida)
    intersectarse = intersec_humano(tablero,coordenadas_barco)
    if (intersectarse == False) and (dentro_rango == True):
        for coordenada in coordenadas_barco:
            tablero[coordenada] = 'O'
        barcos_coordenadas_humano.append(coordenadas_barco)   
        buques.remove(fragata)
        todas_coordenadas = [x for x in todas_coordenadas if x not in coordenadas_barco]



def tablero_humanos(tablero, buques, todas_coordenadas, puntos_cardinales):
    
    while buques != []:

            colocar_barco_humano(tablero, buques, todas_coordenadas, puntos_cardinales)
        
    if (todas_coordenadas == []) and (buques != []):

        tablero_humano(tablero, buques, todas_coordenadas, puntos_cardinales)

    return tablero
tablero_humano = tablero_humanos(tablero_juego, barqui単os, allppartida, puntos_cardinales_def)
```
